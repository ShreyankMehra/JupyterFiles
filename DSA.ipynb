```python
def bubble_sort(li):
    i=0
    n=len(li)
    while i < n-1:
        j=0
        swapped=False
        while j <n-1-i:
            if li[j]>li[j+1]:
                li[j],li[j+1]=li[j+1],li[j]
                swapped=True
            j+=1

        if not swapped:
            break
        i+=1
        
    return "Sorted array : {}".format(li)

def selection_sort(li):
    n=len(li)
    i=0
    while i < n-1:
        j=i+1
        min_index=i
        while j < n:
            if li[j]<li[min_index]:
                
                min_index=j
            j+=1    
        li[i],li[min_index]=li[min_index],li[i]
        i+=1
    return "Sorted array : {}".format(li)    
    
def insertion_sort(li):
    
    n=len(li)
    i=1
    while i < n:
        insert_index=i
        current_value=li[i]
        
        j=i-1
        while j > -1:
            if li[j] > current_value:
                li[j+1]=li[j]
                insert_index=j
                j-=1
            else:
                break
                
        li[insert_index]=current_value
        i+=1
    return "Sorted array : {}".format(li)

def counting_sort(array):
    size=len(array)
    output=[0]*size
    rang=max(array)+1
    count=[0]*(rang)
    
    #counting values
    i=0
    while i < size:
        count[array[i]]+=1
        i+=1
        
    #cumulative sum
    i=1
    while i < rang:
        count[i]+=count[i-1]
        i+=1
        
    #sorting in output array
    i=size-1
    while i >= 0:
        output[count[array[i]]-1]=array[i]
        count[array[i]]-=1
        i-=1
    return "Sorted array : {}".format(output)

def quicksort(arr):
    def partition(array,low,high):
        i=low-1
        j=low

        while j <= high:
            if array[j] <= array[high]:
                i+=1
                array[i],array[j]=array[j],array[i]
            j+=1

        return i

    def quick_sort(array,low=0,high=None):
        if high == None:
            high = len(array)-1

        if low < high:
            pivot=partition(array,low,high)
            quick_sort(array,low,pivot-1)
            quick_sort(array,pivot+1,high)

        return "Sorted array : {}".format(array)
    
    return quick_sort(arr)
```
