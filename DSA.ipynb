```python
def bubble_sort(li):
    i=0
    n=len(li)
    while i < n-1:
        j=0
        swapped=False
        while j <n-1-i:
            if li[j]>li[j+1]:
                li[j],li[j+1]=li[j+1],li[j]
                swapped=True
            j+=1

        if not swapped:
            break
        i+=1
        
    return "Sorted array : {}".format(li)

def selection_sort(li):
    n=len(li)
    i=0
    while i < n-1:
        j=i+1
        min_index=i
        while j < n:
            if li[j]<li[min_index]:
                
                min_index=j
            j+=1    
        li[i],li[min_index]=li[min_index],li[i]
        i+=1
    return "Sorted array : {}".format(li)    
    
def insertion_sort(li):
    
    n=len(li)
    i=1
    while i < n:
        insert_index=i
        current_value=li[i]
        
        j=i-1
        while j > -1:
            if li[j] > current_value:
                li[j+1]=li[j]
                insert_index=j
                j-=1
            else:
                break
                
        li[insert_index]=current_value
        i+=1
    return "Sorted array : {}".format(li)

def counting_sort(array):
    size=len(array)
    output=[0]*size
    rang=max(array)+1
    count=[0]*(rang)
    
    #counting values
    i=0
    while i < size:
        count[array[i]]+=1
        i+=1
        
    #cumulative sum
    i=1
    while i < rang:
        count[i]+=count[i-1]
        i+=1
        
    #sorting in output array
    i=size-1
    while i >= 0:
        output[count[array[i]]-1]=array[i]
        count[array[i]]-=1
        i-=1
    return "Sorted array : {}".format(output)

def quicksort(arr):
    def partition(array,low,high):
        i=low-1
        j=low

        while j <= high:
            if array[j] <= array[high]:
                i+=1
                array[i],array[j]=array[j],array[i]
            j+=1

        return i

    def quick_sort(array,low=0,high=None):
        if high == None:
            high = len(array)-1

        if low < high:
            pivot=partition(array,low,high)
            quick_sort(array,low,pivot-1)
            quick_sort(array,pivot+1,high)

        return "Sorted array : {}".format(array)
    
    return quick_sort(arr)
```

Both merge sort and quick sort are efficient sorting algorithms with different characteristics. Let's compare their time complexities:

    Merge Sort:
        Merge sort has a time complexity of O(n log n) in all cases.
        It divides the array into two halves recursively until each sub-array contains only one element, which takes O(log n) time. Then, it merges the sub-arrays, which takes O(n) time.
        The merging step dominates the time complexity, so the overall time complexity is O(n log n).

    Quick Sort:
        Quick sort has an average-case time complexity of O(n log n).
        However, in the worst-case scenario (e.g., when the pivot is consistently chosen poorly), quick sort's time complexity can degrade to O(n^2).
        The average-case time complexity arises from the partitioning step, which divides the array into two partitions around a pivot element in O(n) time, followed by recursively sorting each partition.
        The worst-case time complexity occurs when the partitioning is unbalanced, resulting in one partition significantly larger than the other, leading to O(n^2) time complexity.

In summary, both merge sort and quick sort have average-case time complexities of O(n log n), but quick sort may perform poorly in some cases due to its worst-case time complexity of O(n^2). Merge sort, on the other hand, maintains a consistent time complexity of O(n log n) regardless of the input array, making it more predictable in terms of performance. However, merge sort may require additional space for the merging step, whereas quick sort is an in-place sorting algorithm, which may be advantageous in terms of memory usage.
